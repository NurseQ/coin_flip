## Coin flipping in python

This notebook is an exploration and use of the binomial distribution to mimic the outcome of flipping coins. 

```python
# import the numerical packages needed for the exercise
import numpy as np
import scipy.stats as ss

# import the packages needed for visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# set the default size of the plots for easier viewing

plt.rcParams['figure.figsize'] = (12 ,8)
```

```python
# (Number of times to flip a coin or n trials, probability, size)
np.random.binomial(1, 0.5, 1)
```

```python
# example of 1000 flips or trials with a 0.5 probability done once
np.random.binomial(1000, 0.5, 1)
```

```python
# calculates the chance that there will be k = 521 for 1000 trials on a 0.5 probability
ss.binom.pmf(521, 1000, 0.5)
```

```python
# displot of a binomial distribution 
sns.distplot(np.random.binomial(1000, 0.5, 1000))
```
